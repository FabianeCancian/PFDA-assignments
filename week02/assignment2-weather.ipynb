```python
import pandas as pd
#Plot
import matplotlib.pyplot as plt
import numpy as np

```

```python
#using Pandas to read the document on csv
df = pd.read_csv('weather.csv')
```

```python
#I output only the information I was going to use in the chart to make a clear visualization. 
df['dewPointTemperature_Celsius']
df['createdDateTime']
df
# I created the variables Temperature and Time to make easy and more understadable the chart code. 
Temperature = df['dewPointTemperature_Celsius'].to_numpy()
print(Temperature,'\n')
Time = df['createdDateTime'].to_numpy()
print(Time,'\n')
```

```python
#To have a clear visualization of the time in the chart I output only the information for hour, I used a substring function, but a convertion from string to datetime and output only the minuts is also possible.
#from datetime import datetime

for index, item in enumerate(Time):
    # Time[index] = datetime.fromisoformat(item).strftime("%H")
    Time[index] = item[11:13]

print(Time,'\n')
```

```python
# Create a new figure and set of axes.
fig, ax = plt.subplots()
#create plot
ax.plot(Time, Temperature)
#create title
ax.set_title('Temperature over time')
#create axis labels
ax.set_xlabel('Last 24 hours') 
ax.set_ylabel('Temperature in Celsius')
```
