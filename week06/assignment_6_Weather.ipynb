```python
# Data frames 

import pandas as pd
import seaborn as sns
```

```python

# Load the dataset.

df = pd.read_csv("https://cli.fusio.net/cli/climate_data/webdata/hly4935.csv", skiprows= 23, low_memory=False)
df.head(10)
```

```python
# Changing the date format from ex: 13-jul-1996 to 1996-07-13.

df['date'] = pd.to_datetime(df['date'], format="mixed")
df['date']
```

```python
#Tranforming th date in index. 

df.set_index('date', inplace=True)
df.head(10)
```

```python
#Plot the total temperature.
sns.lineplot(data=df, x='date', y='temp').set(title='Total Temperature Variance')
```

```python
#Calculate the mean for the month, using the function interpotate to: "estimates values at new time points by finding points along a straight line between existing data points."

month_mean = df.temp.resample('M').mean().interpolate()
```

```python
#Plot

sns.lineplot(month_mean, color=	'#873260').set(title='Mean Temperature per Month')
```

```python
#Calculate the mean temperature for each year.

year_mean = df["temp"].resample("1Y").mean()
```

```python
#Plot

sns.lineplot(data=year_mean, color=	'#362589').set(title='Mean Temperature per Year')
```

```python
#Calculate the mean temperature for each day.

day_mean = df.temp.resample('D').mean().interpolate()
```

```python
#Plot

sns.lineplot(data=day_mean, color=	'#34af91').set(title='Mean Temperature per Day')
```
