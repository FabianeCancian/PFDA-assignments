```python
# Data frames 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# Load the dataset. I used skiproll to skip the legend par of the dataset and be able to load the csv without problems.

df = pd.read_csv("https://cli.fusio.net/cli/climate_data/webdata/hly4935.csv", skiprows= 23, low_memory=False)
df.head(10)
```

```python
# Changing the date format from e.g.: 13-jul-1996 to 1996-07-13.

df['date'] = pd.to_datetime(df['date'], format="mixed")
df['date']
```

```python
# Declaring clear_df as df to be ease to clean the dataset as df we will transform the date colum in index.
clear_df = df
clear_df
```

```python
#Tranforming th date in index. 

df.set_index('date', inplace=True)
df.head(100)
```

```python
#Plot the total temperature.
sns.lineplot(data=df, x='date', y='temp').set(title='Total Temperature Variance')
```

```python
#Calculate the mean for the month, using the function interpotate to: "estimates values at new time points by finding points along a straight line between existing data points."

month_mean = df.temp.resample('M').mean().interpolate()
```

```python
#Plot

sns.lineplot(month_mean, color=	'#873260').set(title='Mean Temperature per Month')
```

```python
#Calculate the mean temperature for each year.

year_mean = df["temp"].resample("1Y").mean()
```

```python
#Plot

sns.lineplot(data=year_mean, color=	'#362589').set(title='Mean Temperature per Year')
```

```python
#Calculate the mean temperature for each day.

day_mean = df.temp.resample('D').mean().interpolate()
```

```python
#Plot

sns.lineplot(data=day_mean, color=	'#34af91').set(title='Mean Temperature per Day')
```

```python
# Using the lambda function replaces any value that is a space (' ') with 0, otherwise it keeps the original value.
# Converting the 'wdsp' column to float data type after applying the lambda function using the function "astype"

clear_df['wdsp'] = clear_df['wdsp'].apply(lambda x: 0 if x == ' ' else x).astype(float)
```

```python
# Plot

plt.figure(figsize=(20, 10))
sns.lineplot(data=clear_df, x='date', y='wdsp', color='#5a0a29').set(title='Total Windspeed')
```

```python
# Calculate the mean Wind speed for each day.

day_mean_wdsp = clear_df.wdsp.resample('D').mean().interpolate()

#Plot

sns.lineplot(data=day_mean_wdsp, color=	'#f08627').set(title='Mean Windspeed per Day')
```

```python
# Create a plot with the wind speed information for the 24 hours on November 30th.

dateFrom = "2024-11-30 00:00:00"   
dateTo = "2024-11-30 23:00:00"
plt.figure(figsize=(8, 4))
sns.lineplot(data=clear_df.loc[dateFrom:dateTo], x="date", y="wdsp").set(title='Wind speed throughout the day on November 30th')
```

```python
# Calculate the max Wind speed for each day.

day_max_wdsp = clear_df.wdsp.resample('D').max().interpolate()

# Plot

sns.lineplot(data=day_max_wdsp, color='#5f099e').set(title='Max Windspeed per Day')
```

```python
# Calculate Monthly Mean of the Daily Max Windspeed.

Montly_max_wdsp = day_max_wdsp.resample('M').mean()

# Plot

sns.lineplot(data=Montly_max_wdsp, color='#191cb6').set(title='Monthly Mean of the Daily Max Windspeeds')
```
