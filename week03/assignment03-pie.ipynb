```python
# Data frames 
import pandas as pd
#Numerical arrays and random numbers
import numpy as np 
#Plot
import matplotlib.pyplot as plt
```

```python
# Load the dataset.

df = pd.read_csv("https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download")
df
```

```python
#Using the function split in a for loop I got the domains for all emails in the column "Email". Using the function value counts I groupped the domains.  
emails = df['Email'].to_numpy()
domains = []

for index, item in enumerate(emails):
    b = item.split("@")
    domains.append(b[1])

df2 =  pd.DataFrame(domains)

count_domains = df2.value_counts()
```

```python
#Plot the pie chart. 
mylabels = ["Example.org", "Example.com", "Example.net"]
mycolors = ["#483D8B", "#FFE4B5", "#8FBC8F"]
plt.pie(count_domains, colors=mycolors, shadow = True)
plt.legend(title = "Email domains:", labels=mylabels)
plt.title("Email domains used by users")
plt.show()
```
